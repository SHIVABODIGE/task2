# Exploratory Data Analysis (EDA)
## Sales Dataset Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('cleaned_sales_dataset.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Age Distribution
plt.figure(figsize=(12,5))
plt.hist(df['Age'], bins=15)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
plt.savefig('age_distribution.png')
plt.close()
```

```python
# Revenue by Category
plt.figure(figsize=(12,5))
df.groupby('Category')['Total_Sales'].sum().plot(kind='bar')
plt.title('Revenue by Category')
plt.show()
plt.savefig('sales_by_category.png')
plt.close()
```

```python
# Revenue by Gender
plt.figure(figsize=(12,4))
df.groupby('Gender')['Total_Sales'].sum().plot(kind='bar')
plt.title('Revenue by Gender')
plt.show()
plt.savefig('sales_by_gender.png')
plt.close()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12,4))
sns.heatmap(df[['Age','Quantity','Unit_Price','Total_Sales']].corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
plt.savefig('correlation_heatmap.png')
plt.close()
```

```python
# Age vs Total Sales
plt.figure(figsize=(12,5))
plt.scatter(df['Age'], df['Total_Sales'])
plt.xlabel('Age')
plt.ylabel('Total Sales')
plt.title('Age vs Total Sales')
plt.show()
plt.savefig('age_vs_sales.png')
plt.close()
```
